```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.rand(100, 1) * 2
```

```python
plt.scatter(X, y)
```

```python
X_ = np.concatenate([np.ones((len(X), 1)), X], 1)
```

```python
theta = np.random.rand(2, 1)
lr = 0.01

grads = []

plt.figure(figsize=(10, 5))
plt.scatter(X, y)

for i in range(100):
    grad = 2/len(X) * X_.T.dot(X_.dot(theta) - y)
    theta = theta - lr * grad
    grads.append(grad)
    
    if i % 10 == 0:
        plt.plot(X, X_.dot(theta))        
```
